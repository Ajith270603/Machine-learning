```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(columns = ['song_name', 'Unnamed: 0', 'title'], inplace = True)
df.head()
```

```python
genres = ['Dark Trap', 'Underground Rap', 'Trap Metal', 'Emo', 'Rap', 'RnB',
       'Pop', 'Hiphop', 'techhouse', 'techno', 'trance', 'psytrance',
       'trap', 'dnb', 'hardstyle']
```

```python
df['genre'].replace({'Dark Trap':0, 'Underground Rap':1, 'Trap Metal':2, 'Emo':3, 'Rap':4, 'RnB':5,
       'Pop':6, 'Hiphop':7, 'techhouse':8, 'techno':9, 'trance':10, 'psytrance':11,
       'trap':12, 'dnb':13, 'hardstyle':14}, inplace = True)
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize = (15,8))
corr= df.corr()
sns.heatmap(corr, cmap ='RdBu' ,annot=True, square = True)
```

```python
X = df.drop(columns = ['type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms'])
y = df['genre']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(f'Length of training set is: {len(X_train)}, length of test set is: {len(X_test)}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
y_train_pred = knn.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

plt.figure(figsize = (15,8))
conf = confusion_matrix(y_train, y_train_pred)
sns.heatmap(conf, annot = True, square = True)
```

```python
print(classification_report(y_train, y_train_pred))
```

```python
y_test_pred = knn.predict(X_test)
print(classification_report(y_test, y_test_pred))
```
